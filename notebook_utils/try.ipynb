```python
import tensorflow.keras.datasets as datasets

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Print the shape of the training and testing data
print("Training data shape:", x_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing data shape:", x_test.shape)
print("Testing labels shape:", y_test.shape)
```

```python
x_train_2 = x_train[y_train == 2]
```

```python
x_train_2.shape
```

```python
x_train_2 = x_train_2.reshape(-1, 28, 28, 1)
```

```python
actual_tra = np.empty((len(x_train_2), 32, 32, 1))
from PIL import Image
from tqdm import tqdm
for i in tqdm(range(len(x_train_2))):
    img = Image.fromarray(x_train_2[i].reshape(28, 28))
    img = img.resize((32, 32))
    np_img = np.array(img)
    actual_tra[i] = np_img.reshape(32, 32, 1)
```

```python
np.save("mnist_db.npy", actual_tra)
```
